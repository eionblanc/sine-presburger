### Example: graphed level sets for removal of linear appearances

$-x + y - 2 < 2\sin(3x + 2y - 1) + \sin(2x)$

```python
L_term = lambda x,y : -x + y - 2
eps = gcd([(1,1),(-1,1)])
cons = L_term(0,0)

osc_Lsin_term = lambda x,y : 2*np.sin(3*x + 2*y - 1) + np.sin(2*x)
R = 3
```

```python
level_sets(eps, cons, R, L_term, osc_Lsin_term, flip=True)
```
